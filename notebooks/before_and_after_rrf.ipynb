```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

plt.style.use('default')
mpl.rcParams.update(mpl.rcParamsDefault)
import numpy as np

before = pd.read_csv("rrf_before_retrieval.csv")
after  = pd.read_csv("rrf_after.csv")
K = 10 


c1 = '#9DBAD5'
c2 = '#4C72B0'



b = before.loc[before["k"] == K].iloc[0]
a = after.loc[after["k"] == K].iloc[0]
metrics = ["recall", "hit", "mrr", "ndcg"]
labels = ["Recall@10", "Hit@10", "MRR@10", "nDCG@10"]

before_vals = [b[m] for m in metrics]
after_vals  = [a[m] for m in metrics]
x = range(len(metrics))
w = 0.38

plt.figure(figsize=(7, 4))
plt.bar([i - w/2 for i in x], before_vals, width=w, label="до", color=c1)
plt.bar([i + w/2 for i in x], after_vals,  width=w, label="после", color=c2)
plt.xticks(list(x), labels)
plt.ylim(0, 1)
plt.ylabel("значение метрики")
plt.xlabel("метрика")
plt.legend(loc="lower right")
plt.tight_layout()
plt.savefig("compare_cachestvo.png", dpi=200)
plt.show()
```
