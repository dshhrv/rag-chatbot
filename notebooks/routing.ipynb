```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("action_accuracy_by_label3.csv")
df = df.sort_values("accuracy_percent", ascending=True)

labels = df["expected_action"]
values = df["accuracy_percent"]

single_color = "#4F76B5"

plt.figure(figsize=(8, 5))
plt.barh(labels, values, color=single_color)

for i, v in enumerate(values):
    plt.text(v + 1, i, f"{v:.1f}%", va="center", fontsize=10)

plt.xlim(0, 100)
plt.xlabel("Accuracy, %")
plt.ylabel("Class")
plt.tight_layout()
plt.savefig("action_accuracy_by_label_single_blue.png", dpi=300, bbox_inches="tight")
plt.show()
```
