```python
import pandas as pd
acc = pd.read_csv('action_accuracy_by_label2.csv')
acc2 = pd.read_csv('action_accuracy_by_label.csv')
```

```python

acc.head(10)
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

plt.style.use('default')
mpl.rcParams.update(mpl.rcParamsDefault)
mpl.rcParams['font.family'] = 'DejaVu Sans'
mpl.rcParams['savefig.bbox'] = 'tight'
mpl.rcParams['savefig.pad_inches'] = 0.3



merged = acc.merge(
    acc2,
    on='expected_action',
    suffixes=('_1', '_2')
)

x = np.arange(len(merged))
width = 0.35

fig, ax = plt.subplots(figsize=(8, 6))

ax.bar(
    x - width / 2,
    merged['accuracy_percent_1'],
    width=width,
    label='до',
    color='#9DBAD5'
)

ax.bar(
    x + width / 2,
    merged['accuracy_percent_2'],
    width=width,
    label='после',
    color='#4C72B0' 
)

ax.set_ylabel('Accuracy (%)')
ax.set_xlabel('Класс')
ax.set_xticks(x)
ax.set_xticklabels(merged['expected_action'], rotation=45, ha='right')

ax.grid(False)
ax.legend()

fig.subplots_adjust(bottom=0.30)

fig.savefig('Accuracy_comparison.png', dpi=300)
plt.show()
```
